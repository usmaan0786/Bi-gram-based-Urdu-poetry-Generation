## Text Generation using bi-grams Model

```python
with open('faiz.txt', 'r', encoding='utf-8') as f:
    verse = f.read()
```

```python
print(verse)
```

```python
import spacy
import random
nlp = spacy.blank('ur')
doc = nlp(verse)
```

```python
doc
```

```python
# sentence tokeization
arr = list()
sentence = ""
for token in doc:
    if token.text=="\n" or token.text=="\n\n" or token.text=="\n\n\n":
        arr.append(sentence)
        sentence = ""
    else:
        sentence += token.text
        sentence +=" "
arr.append(sentence)
print(len(arr))
```

```python
arr
```

```python
# word tokenization
token_word = list()
word = ""
for sentence in arr:
    for letter in sentence:
        if letter != " ":
            word += letter
        else:
            token_word.append(word)
            word = ""
            
# making bigrams
bigrams = []
bigrams = [(token_word[i], token_word[i+1]) for i in range(len(token_word)-1)]

print(bigrams[:10])
```

```python
token_word[:10]
```

```python
list1 = list()
list2 = list()
word = ""
for sentence in arr:
    for letter in sentence:
        if letter != " ":
            word += letter
        else:
            list1.append(word)
            word = ""
    list2.append(list1)
    list1 = []
print(list2)

print("\n")
```

```python
# computing probabilties
result = 0
dic = {}
upper_count = 0
lower_count = 0
prob = 0
for token in bigrams:
    for whole_index in list2:
#         print(whole_index, "\n", token[0], " ", token[1])
        if (token[0] and token[1]) in whole_index:
            upper_count += 1
        if token[0] in whole_index:
            lower_count += 1
#     print(upper_count, lower_count)
    prob = lower_count/upper_count
    result = round(prob, 4)
    upper_count = 0
    lower_count = 0
    dic[token] = result
```

```python
print(dic)
```
